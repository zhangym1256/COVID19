```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')
```

```python
confirmed = pd.read_csv('files/time_series_covid19_confirmed_US.csv')
death = pd.read_csv('files/time_series_covid19_deaths_US.csv')
```

```python
confirmed
```

```python
dates = [x for x in list(confirmed.columns) if re.search('/',x)]
```

```python
confirmed_clean = confirmed.melt(id_vars=['Admin2','Province_State','Country_Region','Lat','Long_','Combined_Key'], value_vars= dates).rename(columns={'Admin2':'County','variable':'Date','value':'Total Cases'})
confirmed_clean['Date'] = pd.to_datetime(confirmed_clean['Date'], errors ='coerce')
confirmed_clean.set_index('Date',inplace = True)
confirmed_clean
```

```python
confirmed_clean.to_csv('US_confirmed.csv')
```

```python
death_clean = death.melt(id_vars=['Admin2','Province_State','Country_Region','Lat','Long_','Combined_Key'], value_vars= dates).rename(columns={'Admin2':'County','variable':'Date','value':'Total Death'})
death_clean['Date'] = pd.to_datetime(death_clean['Date'])
death_clean.set_index('Date',inplace = True)
death_clean
```

```python
death_clean.to_csv('US_death.csv')
```

### NJ New Cases and Death 

```python
NJ_New = confirmed_clean.loc[confirmed_clean['Province_State']=='New Jersey']
```

```python
NJ_New_agg = NJ_New.groupby('Date')[['Total Cases']].sum().reset_index()
NJ_New_agg['New Cases'] =[0]+ [y-x if y >= x else 0 for x,y in zip(NJ_New_agg['Total Cases'][:-1],NJ_New_agg['Total Cases'][1:])]
NJ_New_agg.set_index('Date',inplace = True)
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax2 = ax1.twinx()
ax1.plot(NJ_New_agg.loc[:,'Total Cases'], 'r-')
ax1.tick_params(axis = 'y', labelcolor='r')
ax1.tick_params(axis = 'x', labelrotation = 45)
ax2.plot(NJ_New_agg.loc[:,'New Cases'], 'b--')
ax2.tick_params(axis = 'y',labelcolor='b')

ax1.set_xlabel('Date')
ax1.set_ylabel('Total Cases', color='r')
ax2.set_ylabel('Daily New Cases', color='b')
ax1.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
plt.title('Total and Daily New Cases in NJ',fontsize = 15)
plt.show();
```

```python
NJ_Death = death_clean.loc[death_clean['Province_State']=='New Jersey']
```

```python
NJ_Death_agg = NJ_Death.groupby('Date')[['Total Death']].sum().reset_index()
NJ_Death_agg['New Death'] =[0]+ [y-x if y >= x else 0 for x,y in zip(NJ_Death_agg['Total Death'][:-1],NJ_Death_agg['Total Death'][1:])]
NJ_Death_agg.set_index('Date',inplace = True)
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax2 = ax1.twinx()
ax1.plot(NJ_Death_agg.loc[:,'Total Death'], 'r-')
ax1.tick_params(axis = 'y', labelcolor='r')
ax1.tick_params(axis = 'x', labelrotation=45)
ax2.plot(NJ_Death_agg.loc[:,'New Death'], 'b--')
ax2.tick_params(axis = 'y',labelcolor='b')

ax1.set_xlabel('Date')
ax1.set_ylabel('Total Cases', color='r')
ax2.set_ylabel('Daily New Cases', color='b')
ax1.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
plt.title('Total and Daily New Death in NJ', fontsize = 15)

plt.show();
```
